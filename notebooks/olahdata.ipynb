```python
# library 

import pandas as pd 
```

```python
# load csv dalam data frame
df = pd.read_csv('data.csv')
```

```python
df.info()
```

```python
df.head(50)
```

```python
# ganti semua data ke string

df['detection_timestamp'] = df['detection_timestamp'].astype(str)
```

```python
# ganti value dari 58200 jadi 2026
df['detection_timestamp'] = df['detection_timestamp'].str.replace('58200', '2026')
```

```python
df.head(10)
```

```python
# Konversi 'timestamp' ke datetime terlebih dahulu (ini tanggal yang valid)
df['timestamp'] = pd.to_datetime(df['timestamp'], format='ISO8601')
```

```python
# Ambil TANGGAL dari 'timestamp' (Jadikan string dengan format YYYY-MM-DD)
tanggal_benar = df['timestamp'].dt.strftime('%Y-%m-%d')
```

```python
waktu_benar = df['detection_timestamp'].astype(str).str.split('T').str[1]
```

```python
df['detection_timestamp'] = pd.to_datetime(tanggal_benar + 'T' + waktu_benar, format='ISO8601')
```

```python
df['detection_timestamp'] = df['detection_timestamp'].dt.tz_convert('Asia/Jakarta')
df['timestamp'] = df['timestamp'].dt.tz_convert('Asia/Jakarta')
```

```python
df['detection_timestamp'] = df['detection_timestamp'].dt.tz_localize(None)
df['timestamp'] = df['timestamp'].dt.tz_localize(None)
```

```python
print(df[['detection_timestamp', 'timestamp']].head())
```

```python
df.head(10)
```

```python
df['location'] = df['location'].fillna('demangan')
```

```python
df.head(10)
```

```python
# Kita hitung jumlah track_id yang unik (nunique)
df_harian = df.resample('D', on='detection_timestamp')['track_id'].nunique().reset_index()
```

```python
# 2. Rapikan nama kolomnya agar enak dibaca
df_harian.columns = ['Tanggal', 'Total_Kendaraan']
```

```python
# 3. Urutkan data dari yang paling banyak ke paling sedikit (Descending)
df_harian_tertinggi = df_harian.sort_values(by='Total_Kendaraan', ascending=False)
```

```python
# 4. Tampilkan hasilnya (misal 5 hari paling ramai)
print("--- 5 Hari dengan Volume Kendaraan Tertinggi ---")
print(df_harian_tertinggi.head())
```

```python
df_puncak = df[df['detection_timestamp'].dt.strftime('%Y-%m-%d') == '2026-03-29']
```

```python
print(f"Total baris data di tanggal 29 Maret: {len(df_puncak)}")
df_puncak.head()
```

```python
# Kelompokkan data df_puncak per jam ('H')
df_puncak_per_jam = df_puncak.resample('h', on='detection_timestamp')['track_id'].nunique().reset_index()
df_puncak_per_jam.columns = ['Waktu', 'Volume_Kendaraan']
```

```python
# Tampilkan 5 jam paling sibuk di hari itu
jam_sibuk = df_puncak_per_jam.sort_values(by='Volume_Kendaraan', ascending=False)
print("--- Jam Paling Sibuk di Tanggal 29 Maret 2026 ---")
print(jam_sibuk.head())
```

```python
# Menghitung jenis kendaraan khusus di hari itu
komposisi_kendaraan = df_puncak.groupby('object')['track_id'].nunique().sort_values(ascending=False).reset_index()
print("\n--- Komposisi Kendaraan ---")
print(komposisi_kendaraan)
```
